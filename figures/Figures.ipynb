### Imports and Definitions

```python
from copy import copy
from gzip import open as gopen
from matplotlib import rcParams
from matplotlib.patches import Patch
from pickle import load,dump
from seaborn import pointplot,set_context,set_style
import matplotlib.pyplot as plt
EART = r'E_{ART}'
ED = r'E_{d}'
```

### Aesthetics

```python
RC = {
    "font.size":12,
    "axes.titlesize":16,
    "axes.labelsize":14,
    "legend.fontsize":10,
    "xtick.labelsize":10,
    "ytick.labelsize":10
}

meta = {
    'color': {
        'expected': '#000000',
        'tn93': '#FF0000',
        'tc_max': '#0000AA',
        'tc_max_clade': '#0000FF',
        'tc_avg_clade': '#00FF00'
    },

    'label': {
        'expected': 'Expected',
        'tn93': 'GD + Cluster Growth',
        'tc_max': 'TreeCluster (Max)',
        'tc_max_clade': 'TreeCluster (Max Clade)',
        'tc_avg_clade': 'TreeCluster (Average Clade)'
    },

    'linestyle': {
        'expected': '--',
        'tn93': ':',
        'tc_max': '-',
        'tc_max_clade': '-',
        'tc_avg_clade': '-'
    }
}

set_context("paper", rc=RC)
set_style("ticks")
rcParams['font.family'] = 'serif'
```

### Effectiveness vs. ART

```python
eff_order = ['expected','tn93','tc_max','tc_max_clade','tc_avg_clade']
eff_handles = [Patch(color=meta['color'][m], label=meta['label'][m]) for m in eff_order]
eff_data = load(gopen('DATA.EFFECTIVENESS.pkl.gz'))
eff_ymax = max(e for d in eff_data for r in eff_data[d] for m in eff_data[d][r] for e in eff_data[d][r][m])
rc = copy(RC); rc["legend.fontsize"]=12; rc["xtick.labelsize"]=14; rc["ytick.labelsize"]=14; set_context("paper", rc=rc)
for d in sorted(eff_data.keys()):
    fig = plt.figure()
    for m in eff_order:
        x = list(); y = list()
        for r in sorted(eff_data[d].keys()):
            y += eff_data[d][r][m]; x += [1./r]*len(eff_data[d][r][m])
        ax = pointplot(x=x, y=y, color=meta['color'][m])
    ax.invert_xaxis()
    plt.ylim(ymin=0, ymax=eff_ymax)
    plt.title(r"Effectiveness ($%s=%d$)" % (ED,d), size=22)
    plt.xlabel(r"$%s$ (years)"%EART, size=18)
    plt.ylabel("Average Number of Infections", size=18)
    legend = plt.legend(handles=eff_handles,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., frameon=True)
    plt.show()
    fig.savefig('effectiveness_degree_%s.pdf'%str(d), format='pdf', bbox_extra_artists=(legend,), bbox_inches='tight')
    plt.close()
set_context("paper", rc=RC)
```
